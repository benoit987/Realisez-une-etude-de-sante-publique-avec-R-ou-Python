## 5.4. l’utilisation de la disponibilité intérieure, en particulier la part qui est attribuée à l’alimentation animale, celle qui est perdue et celle qui est concrètement utilisée pour l'alimentation humaine

```python
dispo_alimentaire1
```

```python
#Calculons à partir de la disponibilité intérieure la part qui est attribuée à l’alimentation animale

#Calcul de la somme de la disponibilité intérieure exprimé en tonnes
#Calcul de la somme des aliments pour animaux exprimé en tonnes

total_en_tonnes_disponibilité_intérieur= dispo_alimentaire1['Disponibilité intérieure'].sum()
total_en_tonnes_aliments_pour_animaux=  dispo_alimentaire1['Aliments pour animaux'].sum() 
proportion_dédiée_aux_animaux= total_en_tonnes_aliments_pour_animaux / total_en_tonnes_disponibilité_intérieur

print(proportion_dédiée_aux_animaux)
```

```python
#Calculons à partir de l’utilisation de la disponibilité intérieure la part qui est attribuée aux pertes

#Calcul de la somme des pertes exprimé en tonnes

total_en_tonnes_pertes= dispo_alimentaire1['Pertes'].sum()

proportion_dédiée_aux_pertes= total_en_tonnes_pertes / total_en_tonnes_disponibilité_intérieur
print(proportion_dédiée_aux_pertes)
```

Les pertes représentent 17 % de tonnes de la disponibilité intérieure

```python
#Calculons à partir de la disponibilité intérieure la part qui est utilisée pour l'alimentation humaine

#Calcul de la somme de la part utilisée pour l'alimentation humaine exprimé en tonnes

total_en_tonnes_alimentation_humaine= dispo_alimentaire1['Nourriture'].sum()

proportion_dédiée_aux_humains= total_en_tonnes_alimentation_humaine / total_en_tonnes_disponibilité_intérieur
print(proportion_dédiée_aux_humains)
```

# 6.  Questions de Melanie

 ## 6.1. Les pays pour lesquels la proportion de personnes sous-alimentées est la plus forte en 2017

```python
#Population en état de sous nutrition
population_mondiale_2017_1 = population.loc[population['Année']==2017]
snutrition2017= sous_nutrition.loc[population['Année']==2017]
pop_snutrition2017= pd.merge(snutrition2017,population_mondiale_2017_1, how='left', on='Zone')
```

```python
#On renomme nos colonnes
pop_snutrition2017.rename(columns={'Valeur_x' : 'Personnes en sous-nutrition','Valeur_y': 'Population'}, inplace = True )

#On supprime les colonnes Année_x et Année_y
pop_snutrition2017.drop(columns={'Année_x','Année_y'}, inplace= True)

#Colonne proportion
pop_snutrition2017['Proportion ']=round((pop_snutrition2017['Personnes en sous-nutrition']/pop_snutrition2017['Population']*100),2
                                        )
#On affiche les 10 premiers résultats par ordre décroissant
résultat = pop_snutrition2017.sort_values(by='Proportion ', ascending=False).head(10)
print(résultat)
```

## 6.2.  Les pays qui ont le plus bénéficié d’aide depuis 2013

```python
#On renomme notre colonne 'Valeur' en 'Aides reçus en tonnes)'
aide_alimentaire.rename(columns={'Valeur' : 'Aides reçus en tonnes'}, inplace = True )
```

```python
#Somme des aides reçu en tonnes par pays et par ordre décroissant
resultat2= aide_alimentaire.groupby(['Pays bénéficiaire'])['Aides reçus en tonnes'].sum().sort_values(ascending=False).head(10)
print(resultat2)
```

## 6.3.  Les pays qui ont le plus de disponibilité par habitant en 2017

```python
#somme disponibilité alimentaire par pays (Zone):
pays_forte_disponibilité=dispo_alimentaire1.groupby('Zone').sum('Disponibilité alimentaire (Kcal/personne/jour)').astype(int)
#Les 10 pays avec la plus forte disponibilité alimentaire:
pays_forte_disponibilité=pays_forte_disponibilité.sort_values('Disponibilité alimentaire (Kcal/personne/jour)', ascending=False).head(10)
#On affiche la colonne souhaitée
pays_forte_disponibilité=pays_forte_disponibilité[['Disponibilité alimentaire (Kcal/personne/jour)','Disponibilité alimentaire en quantité (kg/personne/an)']]
print(pays_forte_disponibilité)
```

## 6.4.  Les pays qui ont le moins de disponibilité par habitant en 2017

```python
#somme disponibilité alimentaire par pays (Zone):
pays_faible_disponibilité=dispo_alimentaire1.groupby('Zone').sum('Disponibilité alimentaire (Kcal/personne/jour)').astype(int)
#Les 10 pays avec la plus faible disponibilité alimentaire:
pays_faible_disponibilité=pays_faible_disponibilité.sort_values('Disponibilité alimentaire (Kcal/personne/jour)', ascending=True).head(10)
#On affiche la colonne souhaitée
pays_faible_disponibilité=pays_faible_disponibilité[['Disponibilité alimentaire (Kcal/personne/jour)','Disponibilité alimentaire en quantité (kg/personne/an)']]
print(pays_faible_disponibilité)
```
